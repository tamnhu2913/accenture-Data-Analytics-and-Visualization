# Loading data

```python
content_df = pd.read_csv('/content/drive/MyDrive/Data/accenture Data Analytics/Data/Content.csv', index_col = 0)
reaction_df = pd.read_csv('/content/drive/MyDrive/Data/accenture Data Analytics/Data/Reactions.csv', index_col = 0)
reactype_df = pd.read_csv('/content/drive/MyDrive/Data/accenture Data Analytics/Data/ReactionTypes.csv', index_col = 0)
```

# Cleaning data

## Content dataset

```python
content_df.head()
```

```python
content_df = content_df.rename(columns = {'Type': 'Content Type'})
```

```python
content_df.info()
```

Since, URL and User ID are likely not relevant for understanding which categories are performing the specifically for the content. Therefore, we will delect both comlumns.

```python
content_df = content_df.drop(columns=['URL', 'User ID'])
content_df.info()
```

```python
#Check NULL variable
content_df.isnull().sum()
```

```python
print(content_df['Category'].unique())
```

Through the result above, we can see that the category has many types, but most of them have similar content. For example, in the category 'Studying,' there are two names in this column, such as 'Studying,' and '"studying."' Therefore, we will remove quotation marks (") and convert to lowercase for data synchronization.

```python
content_df['Category'] = content_df['Category'].str.replace('"', '')
content_df['Category'] = content_df['Category'].str.lower()
print(content_df['Category'].unique())
```

## Reaction dataset

```python
reaction_df.head()
```

```python
reaction_df = reaction_df.rename(columns = {'Type': 'Reaction Type'})
```

```python
reaction_df.info()
```

Change type of Datetime column into date

```python
reaction_df['Datetime'] = reaction_df['Datetime'].apply(pd.to_datetime)
reaction_df.info()
```

Similarly, we will delete User ID column.

```python
reaction_df = reaction_df.drop(columns=['User ID'])
reaction_df.head()
```

```python
reaction_df.isnull().sum()
```

We see that Type have NaN variable. So we will delect the rows have NaN variable.

```python
reaction_df = reaction_df.dropna(subset=['Reaction Type'])
reaction_df.shape
```

## Reaction Type dataset

```python
reactype_df.head()
```

```python
reactype_df = reactype_df.rename(columns = {'Type': 'Reaction Type'})
```

```python
#Type of each columns
reactype_df.info()
```

```python
#check Null value
reactype_df.isnull().sum()
```

We don't need change anything in Reaction Type dataset.

# Data Modeling

## Merge 3 dataset

Use Reaction as a base dataset.

```python
merge_df = pd.merge(reaction_df, content_df, on = 'Content ID')
merge_df = pd.merge(merge_df, reactype_df, on = 'Reaction Type')
merge_df.head()
```

```python
merge_df.to_csv('cleaning_data.csv')
```

## Top 5 performing categories

```python
count_category = merge_df.groupby('Category')['Score'].sum().sort_values(ascending = False).reset_index()
count_category = count_category.rename(columns = {'Score': 'Total Score'})
count_category[:5]
```
